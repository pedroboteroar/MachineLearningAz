## The Machine Leaning Process

1. Data Preprocessing: Import, clena and split the data into training and test sets.
2. Modelling: Build the model, train the model, make predictions.
3. Evaluation: Calculate performance metrics and make a verdict on the model.

## Splitting a dataset into a training set and a test set

Splitting data into a training set and a test set is a common operation in machine learning. The training set is used to train the model and the test set is used to evaluate the model. The test set is used to evaluate the model because it is data that the model has not seen before. If the model performs well on the test set, we can be confident that it will perform well on unseen data.

Usually, the training set is larger than the test set. A common split is 80% training data and 20% test data.

## Feature Scaling

It is always applied to columns, not applied across columns. 

Feature scaling are many techniques used to change the way in which we present the data. The two most significant techniques are standardisation and normalisation.

### Standardisation

Standardisation is the process of rescaling the features so that they have the properties of a standard normal distribution with a mean of zero and a standard deviation of one.

The formula for standardisation is:

$$x_{stand} = \frac{x - \mu}{\sigma}$$

where $x$ is the original feature vector, $\mu$ is the mean of that feature vector and $\sigma$ is its standard deviation.

### Normalisation

Normalisation is the process of rescaling the features so that they have the properties of a normal distribution with a mean of zero and a standard deviation of one.

The formula for normalisation is:

$$x_{norm} = \frac{x - x_{min}}{x_{max} - x_{min}}$$

where $x$ is the original feature vector, $x_{min}$ is the smallest value in that feature vector and $x_{max}$ is the largest value in that feature vector.

### Example

Supposing we have a dataframe of two columns which include the age and salary of a person such as:

| Age | Salary |
| --- | ------ |
| 45  | 70000  |
| 44  | 60000  |
| 40  | 52000  |

We would see that the difference between the age and salary is very large. This is because the age is in the range of 40 to 45 and the salary is in the range of 52000 to 70000. This is a problem because the model will give more weight to the salary because it has a larger range. To solve this problem, we can use feature scaling.

If we apply standardisation to the age and salary columns, we would get:

| Age      | Salary   |
| -------- | -------- |
| 0.707107 | 1.212678 |
| 0.565685 | 0.212678 |
| -1.27279 | -1.42536 |

To see how these values were calculated, we can use the formula for standardisation:

$$x_{stand} = \frac{x - \mu}{\sigma}$$

For the age column, we would get:

$$x_{stand} = \frac{45 - 43}{2} = 0.707107$$

For the salary column, we would get:

$$x_{stand} = \frac{70000 - 60666.666666666664}{8660.254037844387} = 1.212678$$

If we apply normalisation to the age and salary columns, we would get:

| Age      | Salary   |
| -------- | -------- |
| 1        | 1        |
| 0.8      | 0.4      |
| 0        | 0        |

To see how these values were calculated, we can use the formula for normalisation:

$$x_{norm} = \frac{x - x_{min}}{x_{max} - x_{min}}$$

For the age column, we would get:

$$x_{norm} = \frac{45 - 40}{45 - 40} = 0.75$$

For the salary column, we would get:

$$x_{norm} = \frac{70000 - 52000}{70000 - 52000} = 1$$

#### Results

Using these new tables, we can see how the first two persons are more similar in terms of age and salary. This is because the values are closer together. This is the result of feature scaling.
